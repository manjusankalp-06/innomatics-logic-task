# Problem 1: Employee Performance Bonus Elgibility

```python
employees = {
"Ravi": 92,
"Anita": 88,
"Kiran": 92,
"Suresh": 85
}
highest_score = max(employees.values())
top_performers = []

for name, score in employees.items():
    if score == highest_score:
        top_performers.append(name)


print("Top Performers Eligible for Bonus:",
      ", ".join(top_performers),
      f"(Score: {highest_score})")
```

# Problem 2: Search Query Keyword Analysis

```python
query = "Buy moblie phone buy phone online"
query = query.lower()
words = query.split()
count = {}

for word in words:
    if word in count:
        count[word] = count[word] + 1
    else:
        count[word] = 1

result = {}

for word in count:
    if count[word] > 1:
        result[word] = count[word]

print(result)
```

# Problem 3: Sensor Data Validation

```python
sensor_readings = [3, 4, 7, 8, 10, 12, 5]
valid_readings = []

for i in range(len(sensor_readings)):
    if sensor_readings[i] % 2 == 0:
        valid_readings.append((i, sensor_readings[i]))

print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

# Problem 4: Email Domain Usage Analysis

```python
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

domain_count = {}

for email in emails:
    domain = email.split("@")[1]

    if domain in domain_count:
        domain_count[domain] = domain_count[domain] + 1
    else:
        domain_count[domain] = 1

total = len(emails)

for domain in domain_count:
    percentage = (domain_count[domain] / total) * 100
    print(domain + ":", str(int(percentage)) + "%")
```

# Problem 5: Sales Spike Detection

```python
sales = [1200, 1500, 900, 2200, 1400, 3000]
total = 0

for s in sales:
    total = total + s

average = total /len(sales)

threshold = average + (0.30 * average)

for i in range(len(sales)):
    if sales[i] > threshold:
        print("Day", i + 1, ":", sales[i])
```

# Problem 6: Duplicate User ID Detection

```python
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]
count = {}

for user in user_ids:
    if user in count:
        count[user] = count[user] + 1
    else:
        count[user] = 1
        
for user in count:
    if count[user] > 1:
        print(user + " -> " + str(count[user]) + " times")
```
