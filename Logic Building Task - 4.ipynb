# Problem 1: Mobile Recharge Validation System

```python
def validate_recharge():
    valid_plans = [199, 299, 399, 599]

    while True:
        amount = int(input("Enter recharge amount:"))

        if amount < 50:
            print("Recharge amount must be at least 50")
        elif amount not in valid_plans:
            print("Invalid plan selected")
        else:
            print("Recharge Successful")
            break

validate_recharge()
```

# Problem 2: Inventory Reorder Alert System

```python
def check_inventory(products):
    for item in products:
        if products[item] < 15:
            print(item, ": Reorder Alert")
        else:
            print(item, ": Stock OK")

products = {
    "Rice": 10,
    "Sugar": 25,
    "Oil": 12,
    "Salt": 30
}

check_inventory(products)
```

# Problem 3: Student Result Processing System

```python
def check_result(marks):
    total = 0

    for m in marks:
        total = total + m

    average = total / len(marks)

    print("Average Marks:", average)

    if average >= 50:
        return "Pass"
    else:
        return "Fail"

marks = [60, 55, 40, 70]
result = check_result(marks)
print("Result:", result)
```

# Problem 4: Cab Fare Estimator with Retry

```python
def calculate_fare(distance, peak):
    base_fare = 50
    fare = base_fare + (12 * distance)

    if peak == "yes":
        fare = fare + (0.25 * fare)

    return fare

while True:
    km = int(input("Enter distance in km: "))
    peak_hour = input("Is it peak hour? (yes/no): ")

    total_fare = calculate_fare(km, peak_hour)

    print("Total Fare:", total_fare)

    retry = input("Do you want to calculate again? (yes/no): ")
    if retry != "yes":
        break
```

# Problem 5: Employee Attendance Eligibility Checker

```python
def check_attendance(attendance):
    present = 0

    for day in attendance:
        if day == "P":
            present = present + 1

    percentage = (present / len(attendance)) * 100

    print("Attendance Percentage:", percentage)

    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"

attendance = ["P", "P", "A", "P", "P"]

status = check_attendance(attendance)
print("Status:", status)
```

# Problem 6: Password Strength Checker

```python
def check_password(password):
    has_digit = False
    has_special = False

    if len(password) < 8:
        return "Weak Password"

    for ch in password:
        if ch.isdigit():
            has_digit = True
        if ch in "@#$":
            has_special = True

        if has_digit and has_special:
            return "Strong Password"
        else:
            return "Weak Password"

password = input("Enter password:")
print(check_password(password))
```
