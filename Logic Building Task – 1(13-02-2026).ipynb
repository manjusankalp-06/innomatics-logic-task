```python
# Message Length Analyzer

message = "Hi", "Welcome to the platform"

length = len(message)

print("Length:", length)

if length > 10:
    print("Message is longer than 10 characters")
else:
    print("Message is short")
```

```python
# Error Message Detector

logs = ["INFO", "ERROR", "WARNING", "ERROR"]

count = 0

for log in logs:
    if log == "ERROR":
        count = count + 1

print("Total ERROR:", count)
```
