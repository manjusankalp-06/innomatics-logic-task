# Problem 1: Smart Parking Lot Management System

```python
def manage_parking(capacity, logs):
    current = 0
    peak = 0

    for log in logs:
        if log == "IN":
            current = current + 1
        elif log == "OUT":
            current = current - 1

        # Track peak usage
        if current > peak:
            peak = current

    print("Currently Parked Vehicles:", current)

    if current > capacity:
        print("Parking Status: Over Capacity")
    else:
        print("Parking Status: Available")

    print("Peak Parking Usage:", peak)


capacity = 50
vehicle_logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

manage_parking(capacity, vehicle_logs)
```

# Problem 2: Online Food Delivery Time Estimator

```python
def estimate_delivery_time(distance, traffic, weather):
    base_time = distance * 5   # 5 minutes per km

    if traffic == "High":
        base_time = base_time + 15
    elif traffic == "Medium":
        base_time = base_time + 10

    if weather == "Rainy":
        base_time = base_time + 10

    print("Estimated Delivery Time:", base_time, "minutes")


estimate_delivery_time(8, "High", "Rainy")
```

# Problem 3: Seat Occupancy Analyzer

```python
def analyze_occupancy(total_seats, booked_seats):
    booked = len(booked_seats)
    percentage = (booked / total_seats) * 100

    print("Occupancy:", int(percentage), "%")

    if percentage == 100:
        print("Show Status: Housefull")
    elif percentage >= 75:
        print("Show Status: Almost Full")
    else:
        print("Show Status: Seats Available")


total_seats = 200
booked_seats = [1] * 150

analyze_occupancy(total_seats, booked_seats)
```

# Problem 4: Server Load Classification

```python
def classify_server_load(cpu_readings):
    total = 0

    for reading in cpu_readings:
        total = total + reading

    average = total / len(cpu_readings)

    print("Average CPU Load:", int(average), "%")

    if average < 50:
        print("Server Status: Normal")
    elif average <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")


cpu_data = [45, 60, 70, 85, 90]

classify_server_load(cpu_data)
```

# Problem 5: Smart Classroom Resource Usage Monitor

```python
def monitor_resources(usage):
    overused = []

    for resource in usage:
        if usage[resource] > 8:
            overused.append(resource)

    if len(overused) > 0:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")


resource_usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

monitor_resources(resource_usage)
```

# Problem 6: Online Event Registration Capacity Controller

```python
def manage_registrations(capacity, registrations):
    if registrations > capacity:
        confirmed = capacity
        waitlist = registrations - capacity
        status = "Closed"
    else:
        confirmed = registrations
        waitlist = 0
        status = "Open"

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlist)
    print("Registration Status:", status)


manage_registrations(100, 105)
```
